### Implimentation of K-Mean Clustering

#### Helper Funtions

```python
#your code here
def random_centroids_initializer(data, k):
    # Create an empty list to store the centroids
    centroids = []
    
    # Randomly select k data points as centroids
    random_indexes = random.sample(range(data.shape[0]), k)
    
    # Assign the randomly selected data points as centroids
    for index in random_indexes:
        centroids.append(data[index])
    
    return np.array(centroids)
```

```python
#your code here
def assign_data_points_to_nearest_centroid(data, centroids):
    assigned_centroids = []
    
    # Iterate over each data point
    for point in data:
        # Calculate the distances to each centroid
        distances = [np.linalg.norm(point - centroid) for centroid in centroids]
        
        # Find the index of the nearest centroid
        nearest_centroid = np.argmin(distances)
        
        # Assign the data point to the nearest centroid
        assigned_centroids.append(nearest_centroid)
    
    return np.array(assigned_centroids)
```

```python
#your code here

def update_centroids_by_mean(data, assigned_centroids, k):
    centroids = np.empty((k, data.shape[1]))
    
    # Iterate over each centroid
    for i in range(k):
        assigned_points = data[np.where(assigned_centroids == i)[0]]
        
        # Check if any data points are assigned to the centroid
        if len(assigned_points) > 0:
            # Calculate the mean of assigned data points
            centroid = np.mean(assigned_points, axis=0)
            
            # Assign the updated centroid
            centroids[i] = centroid
        else:
            # If no data points are assigned, keep the previous centroid
            centroids[i] = data[i]
    
    return centroids
```

#### Mian Function

```python
import random
def kmeans_clustering(data, k, max_iterations=100):
    # Randomly initialize centroids
    centroids = random_centroids_initializer(data, k)
    
    for _ in range(max_iterations):
        # Assign each data point to the nearest centroid
        assigned_centroids = assign_data_points_to_nearest_centroid(data, centroids)
        
        # Update centroids by taking the mean of the assigned data points
        new_centroids = update_centroids_by_mean(data, assigned_centroids, k)
        
        # Check convergence
        if np.array_equal(centroids, new_centroids):
            break
        
        # Update centroids to the latest centroids
        centroids = new_centroids

    return assigned_centroids, centroids
```

### K-Mean Clustering on the all images

NOTE: IF YOUR CODE IS TAKING TOO MUCH TIME TO RUN, TRY RESIZE THE IMAGE TO SMALLER SIZE 

```python
k_values = [2, 5, 10,20]
images = [input_img_1_2d, input_img_2_2d, input_img_3_2d]
org_images = [input_img_1, input_img_2, input_img_3]

for i, (image, org_image) in enumerate(zip(images, org_images)):
    for k in k_values:
        labels, centroids = kmeans_clustering(image, k)
        # Replace pixel values with centroid values
        new_image_data = centroids[labels].reshape(org_image.shape)
        # Convert the data back to image format
        new_image = Image.fromarray(new_image_data.astype(np.uint8))
        # Display the new image
        display(HTML(f'<h2>Clustered Image of input {i+1} (k={k})</h2>'))
        display(new_image)
```
