 ### Fit Linear Regression Model (Training data)

 As our linear Model was
 
![Cost Funtion](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Cost%20funtion1.png?raw=true
)

And using derivatives we transformed our model into 2 simultaneous equations

![Model Equations](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Linear%20Model%20Eqs.png?raw=true)


Then converted it in matrix form

![Model Equations](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Linear%20Model.png?raw=true)



NOTE :  Now you are required to compute values wihtout using foor loop on tarining data
- HINT : Use numpy library for this purpose 

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 
matrix_A = np.array([[x_train.shape[0], sum(x_train)], 
                     [sum(x_train), np.sum(np.power(x_train, 2))]])
print(f"Matrix A:\n\n{matrix_A}")

matrix_B = np.array([[sum(y_train)], 
                     [np.dot(y_train, x_train)]])
print(f"\nMatrix B:\n\n{matrix_B}")
```

Expected output

![Linear Model MAtrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20Linear.png?raw=true)


#### Compute Valus of Both Θ`s


```python
#Your code Here
#Calculate the values of 0_node and 0_1 using Matrix Multiplication
#Hint: Use X = A^-1 * B to calculate the values of 0_node and 0_1
X = np.dot(np.linalg.inv(matrix_A), matrix_B)
print(f"Matrix X for the linear model :\n\n{X}")
```

Expected Output

![Linear Model MAtrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20LinearT.png?raw=true)


### Run Predictions (Testing data)

```python
# Your code Here
# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x
# _____DO NOT USE FOR LOOP_____
# instead Convert the column of test_x into numpy a 2 dimensional matrix
# then use matrix multiplication to calculate the predicted values of y
y_pred = X[0][0] + np.multiply(X[1][0], x_test)
```

### Mean Square Error

![Cost Funtion](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Cost%20funtion1.png?raw=true
)

```python
#Your code Here
#Using cost function calculate the cost of the model
cost = (sum((y_pred - y_test)**2)) / (y_test.shape[0]) 
print(f"Mean Square Error for LINEAR MODEL is: {cost}")
```

Expected Output

![MSE of Linear](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/MSE%20Linear.png?raw=true
)

### Plot Data (Train and Test Data)

```python
##Your code Here
#Plot the predicted values of y against the test_x
plt.scatter(x_train, y_train, color = "purple")
plt.plot(x_test, y_pred ,color = "red")
plt.legend(("Train Data", "Linear model"))
plt.title("Linear Model for the train data")
plt.xlabel("X")
plt.ylabel("R")
plt.show()
plt.scatter(x_test, y_test, color = "purple")
plt.plot(x_test, y_pred ,color = "red")
plt.legend(("Test Data", "Linear model"))
plt.title("Linear Model for the test data")
plt.xlabel("X")
plt.ylabel("R")
plt.show()
```

## Now Fit The Quadratic Model

And using derivatives we transformed our model into 3 simultaneous equations

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 

A = np.array([[x_train.shape[0], sum(x_train), np.sum(np.power(x_train, 2))], 
              [sum(x_train), np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3))],
              [np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4))]])
print(f"Matrix A :\n\n{A}")

B = np.array([[sum(y_train)], 
              [np.dot(y_train, x_train)], 
              [np.dot(y_train, np.square(x_train))]])
print(f"\nMatrix B :\n\n{B}")

```

## Now Fit The 4 degree Model

```python
#Your code Here
np.set_printoptions(precision=2) #formating numpy array values

#Finding A and B matrices
A = np.array([[x_train.shape[0], sum(x_train), np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4))], 
              [sum(x_train), np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5))],
              [np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6))],
              [np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7))],
              [np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7)), np.sum(np.power(x_train, 8))]])
print(f"Matrix A :\n\n{A}")

B = np.array([[sum(y_train)], 
              [np.dot(y_train, x_train)], 
              [np.dot(y_train, np.square(x_train))], 
              [np.dot(y_train, np.power(x_train, 3))],
              [np.dot(y_train, np.power(x_train, 4))]])
print(f"\nMatrix B :\n\n{B}")

#Calculate the values of 0_node, 0_1, 0_2, 0_3, 0_4 using Matrix Multiplication
X = np.dot(np.linalg.inv(A), B)
print(f"Matrix X for the 4 degree model :\n\n{X}")

# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x + 0_2 * x2 + 0_3 * x3 + 0_4 * x4
y_pred = X[0][0] + np.dot(X[1][0], x_test) + np.dot(X[2][0], np.square(x_test)) + np.dot(X[3][0], np.power(x_test, 3)) + np.dot(X[4][0], np.power(x_test, 4))

#Using cost function calculate the cost of the model
cost = (sum((y_pred - y_test)**2)) / (y_test.shape[0]) 
print(f"Mean Square Error: {cost}")

#Plot the predicted values of y against the test_x
plt.scatter(x_train, y_train, color = "purple")
plt.plot(x_test, y_pred ,color = "red")
plt.legend(("Train Data", "4 degree model"))
plt.title("4 degree Model for the train data")
plt.xlabel("X")
plt.ylabel("R")
plt.show()

plt.scatter(x_test, y_test, color = "purple")
plt.plot(x_test, y_pred ,color = "red")
plt.legend(("Test Data", "4 degree model"))
plt.title("4 degree Model for the test data")
plt.xlabel("X")
plt.ylabel("R")
plt.show()





















```

```python
#Your code Here
np.set_printoptions(precision=2) #formating numpy array values

#Finding A and B matrices
A = np.array([[x_train.shape[0], sum(x_train), np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6))], 
              [sum(x_train), np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7))],
              [np.sum(np.power(x_train, 2)), np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7)), np.sum(np.power(x_train, 8))],
              [np.sum(np.power(x_train, 3)), np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7)), np.sum(np.power(x_train, 8)), np.sum(np.power(x_train, 9))],
              [np.sum(np.power(x_train, 4)), np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7)), np.sum(np.power(x_train, 8)), np.sum(np.power(x_train, 9)), np.sum(np.power(x_train, 10))],
              [np.sum(np.power(x_train, 5)), np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7)), np.sum(np.power(x_train, 8)), np.sum(np.power(x_train, 9)), np.sum(np.power(x_train, 10)), np.sum(np.power(x_train, 11))],
              [np.sum(np.power(x_train, 6)), np.sum(np.power(x_train, 7)), np.sum(np.power(x_train, 8)), np.sum(np.power(x_train, 9)), np.sum(np.power(x_train, 10)), np.sum(np.power(x_train, 11)), np.sum(np.power(x_train, 12))]])
print(f"Matrix A :\n\n{A}")

B = np.array([[sum(y_train)], 
              [np.dot(y_train, x_train)], 
              [np.dot(y_train, np.square(x_train))], 
              [np.dot(y_train, np.power(x_train, 3))],
              [np.dot(y_train, np.power(x_train, 4))],
              [np.dot(y_train, np.power(x_train, 5))],
              [np.dot(y_train, np.power(x_train, 6))]])
print(f"\nMatrix B :\n\n{B}")

#Calculate the values of 0_node, 0_1, 0_2, 0_3, 0_4, 0_5 and 0_6 using Matrix Multiplication
X = np.dot(np.linalg.inv(A), B)
print(f"Matrix X for the 6 degree model :\n\n{X}")

# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x + 0_2 * x2 + 0_3 * x3 + 0_4 * x4 + 0_5 * x5 + 0_6 * x6
y_pred = X[0][0] + np.dot(X[1][0], x_test) + np.dot(X[2][0], np.square(x_test)) + np.dot(X[3][0], np.power(x_test, 3)) + np.dot(X[4][0], np.power(x_test, 4)) + np.dot(X[5][0], np.power(x_test, 5)) + np.dot(X[6][0], np.power(x_test, 6))

#Using cost function calculate the cost of the model
cost = (sum((y_pred - y_test)**2)) / (y_test.shape[0]) 
print(f"Mean Square Error: {cost}")

#Plot the predicted values of y against the test_x
plt.scatter(x_train, y_train, color = "purple")
plt.plot(x_test, y_pred ,color = "red")
plt.legend(("Train Data", "6 degree model"))
plt.title("6 degree Model for the train data")
plt.xlabel("X")
plt.ylabel("R")
plt.show()

plt.scatter(x_test, y_test, color = "purple")
plt.plot(x_test, y_pred ,color = "red")
plt.legend(("Test Data", "6 degree model"))
plt.title("6 degree Model for the test data")
plt.xlabel("X")
plt.ylabel("R")
plt.show()





















```
