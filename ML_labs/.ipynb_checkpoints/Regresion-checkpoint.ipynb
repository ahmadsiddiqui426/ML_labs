## Now Fit The Quadratic Model

 As our Quadratic Model was
 
![Cost Funtion for quadratic ](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Cost%20funtion2.png?raw=true
)

And using derivatives we transformed our model into 3 simultaneous equations

![Model Equations of quadratic](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Quad%20Model%20Eqs.png?raw=true)

NOTE :  Now you are required to compute values wihtout using foor loop on tarining data
- HINT : Use numpy library for this purpose 

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x
matrix_A = np.array([[x_train.shape[0], sum(x_train), np.dot(x_train, x_train)], 
                    [sum(x_train), np.dot(x_train, x_train), np.power(x_train, 3).sum()], 
                    [np.power(x_train, 2).sum(), np.power(x_train, 3).sum(), np.power(x_train, 4).sum()]])

print(f"Matrix A:\n\n{matrix_A}")

matrix_B = np.array([[sum(y_train)], [np.dot(y_train, x_train)], [np.dot(y_train, np.power(x_train, 2))]])
print(f"\nMatrix B:\n\n{matrix_B}")
```

```python
#Your code Here
#Calculate the values of 0_node and 0_1 using Matrix Multiplication
#Hint: Use X = A^-1 * B to calculate the values of 0_node , 0_1 and 0_2
# Calculate the inverse of matrix A
matrix_A_inv = np.linalg.inv(matrix_A)
theta = np.dot(matrix_A_inv, matrix_B)
theta_matrix = np.matrix(theta)
print("Matrix X for the quadratic model:\n{}".format(theta_matrix))
```

- HINT : Use numpy library for this purpose 

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x 
matrix_A = np.array([[x_train.shape[0], x_train.sum(), np.dot(x_train, x_train), np.dot(np.power(x_train, 2), x_train)],
                     [x_train.sum(), np.dot(x_train, x_train), np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train)],
                     [np.dot(x_train, x_train), np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train), np.dot(np.power(x_train, 4), x_train)],
                     [np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train), np.dot(np.power(x_train, 4), x_train), np.dot(np.power(x_train, 5), x_train)]])
print(f"Matrix A:\n\n{matrix_A}")
matrix_B = np.array([[y_train.sum()], [np.dot(y_train, x_train)], [np.dot(y_train, np.power(x_train, 2))], [np.dot(y_train, np.power(x_train, 3))]])
print(f"Matrix B:\n\n{matrix_B}")
```

## Now Fit The 4 degree Model

```python
#Your code Here
#Now Fit The 4 degree Model
# Define matrix A using numpy functions
matrix_A = np.array([[x_train.shape[0], x_train.sum(), np.dot(x_train, x_train), np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train)],
                     [x_train.sum(), np.dot(x_train, x_train), np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train), np.dot(np.power(x_train, 4), x_train)],
                     [np.dot(x_train, x_train), np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train), np.dot(np.power(x_train, 4), x_train), np.dot(np.power(x_train, 5), x_train)],
                     [np.dot(np.power(x_train, 2), x_train), np.dot(np.power(x_train, 3), x_train), np.dot(np.power(x_train, 4), x_train), np.dot(np.power(x_train, 5), x_train), np.dot(np.power(x_train, 6), x_train)],
                     [np.dot(np.power(x_train, 3), x_train), np.dot(np.power(x_train, 4), x_train), np.dot(np.power(x_train, 5), x_train), np.dot(np.power(x_train, 6), x_train), np.dot(np.power(x_train, 7), x_train)]])
print(f"Matrix A:\n\n{matrix_A}")

# Define matrix B using numpy functions
matrix_B = np.array([[y_train.sum()], [np.dot(y_train, x_train)], [np.dot(y_train, np.power(x_train, 2))], [np.dot(y_train, np.power(x_train, 3))], [np.dot(y_train, np.power(x_train, 4))]])
print(f"Matrix B:\n\n{matrix_B}")

# Calculate the coefficients using matrix multiplication
matrix_A_inv = np.linalg.inv(matrix_A)
theta = np.dot(matrix_A_inv, matrix_B)
theta_matrix = np.matrix(theta)
print("Matrix X for the 4-degree model:\n{}".format(theta_matrix))

# Calculate the predicted values of y for the test data using the coefficients
test_x = testData['X']
test_x_2 = np.power(test_x, 2)
test_x_3 = np.power(test_x, 3)
test_x_4 = np.power(test_x, 4)
test_y_actual = testData['R']

theta_0 = theta[0][0]
theta_1 = theta[1][0]
theta_2 = theta[2][0]
theta_3 = theta[3][0]
theta_4 = theta[4][0]
test_y_pred = theta_0 + (theta_1 * test_x) + (theta_2 * test_x_2) + (theta_3 * test_x_3) + (theta_4 * test_x_4)

# Calculate the mean square error of the model
mse = np.sum((test_y_pred - test_y_actual)**2) / len(test_y_actual)
print("Mean Square Error for the Cubic Model: {}".format(mse))

#Plot the predicted values of y against the test_x
plt.plot(trainData['X'], trainData['R'], 'o',label='Train Data',color='pink')
plt.plot(test_x, test_y_pred, label='4th Degree Model',color='green')
plt.xlabel('X')
plt.ylabel('R')
plt.title('Training Data')
plt.legend()
plt.show()

plt.plot(testData['X'], testData['R'], 'o',label='Test Data',color='purple')
plt.plot(test_x, test_y_pred, label='4th Degree Model',color='green')
plt.xlabel('X')
plt.ylabel('R')
plt.title('Test Data')
plt.legend()
plt.show()
```
